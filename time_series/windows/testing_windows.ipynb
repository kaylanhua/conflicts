```python
import matplotlib.pyplot as plt

def plot_mse_vs_window_size(country, r=11):
    filename = '../../data/ucdp/' + country + '.csv'
    window_sizes = range(1, r)
    mse_results = []
    
    for size in window_sizes:
        window_rf = WindowRandomForest(filename, window_size=size)
        window_rf.load_data()
        mse = window_rf.compute_MSE()
        mse_results.append(mse)
    
    plt.plot(window_sizes, mse_results, marker='o')
    plt.title('MSE vs. Window Size for ' + country)
    plt.xlabel('Window Size')
    plt.ylabel('MSE')
    plt.show()

plot_mse_vs_window_size(country)
```
